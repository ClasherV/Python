```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
dataset=pd.read_csv("Datasets/loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
ms=MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_ms"]=pd.DataFrame(ms.transform(dataset[["CoapplicantIncome"]]),columns=["x"])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_ms"])
plt.show()
```
