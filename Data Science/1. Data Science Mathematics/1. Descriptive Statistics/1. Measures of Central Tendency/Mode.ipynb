```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("Datasets/Titanic-Dataset.csv")
```

```python
dataset["Fare"].mode()
```

```python
type(dataset["Fare"].mode())
```

```python
dataset["Fare"].mode()[0]
```

```python
dataset["Fare"].value_counts()
```

```python
mn=dataset["Fare"].mean()
md=dataset["Fare"].median()
mo=dataset["Fare"].mode()[0]
```

```python
sns.histplot(x="Fare",data=dataset,bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range(0,300)],[i for i in range(0,300)],c="r")
plt.plot([md for i in range(0,300)],[i for i in range(0,300)],c="b")
plt.plot([mo for i in range(0,300)],[i for i in range(0,300)],c="g")
plt.show()
```

```python
sns.histplot(x="Fare",data=dataset,bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range(0,300)],[i for i in range(0,300)],c="b",label="Mean")
plt.plot([md for i in range(0,300)],[i for i in range(0,300)],c="g",label="Median")
plt.plot([mo for i in range(0,300)],[i for i in range(0,300)],c="r",label="Mode")
plt.legend()
plt.show()
```
