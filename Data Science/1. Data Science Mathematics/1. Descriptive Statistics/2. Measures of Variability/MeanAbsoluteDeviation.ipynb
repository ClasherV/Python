```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sec_A=np.array([75,65,73,68,72,67])
sec_B=np.array([90,47,43,96,93,51])
num=np.array([1,2,3,4,5,6])
```

```python
mean=np.mean(sec_A)
```

```python
plt.figure(figsize=(10,3))
plt.scatter(sec_A,num,color="r",label="Section A")
plt.scatter(sec_B,num,color="g",label="Section B")
plt.plot([70,70,70,70,70,70],num,color="b",label="Mean")
plt.legend()
plt.show()
```

```python
mad_a=np.sum(np.abs(sec_A-mean))/len(sec_A)
mad_b=np.sum(np.abs(sec_B-mean))/len(sec_B)
```
